# *Notebook* y *Markdown*

# El *Notebook*

El *Notebook* *Jupyter* es una aplicación web (aunque se puede correr *offline*) que permite crear miniprogramas e interactuar con ellos. Está formado de **celdas**.

Hay esencialmente 2 tipos de celda:

- **Código**. Contienen un miniprograma.

```julia
# Celda de código
1+2+3+4
```

- **Texto**. Contienen texto que se formatea con *Markdown*.

## Celda de texto

texto, *texto*, texto! $A\subset B $

Las celdas pueden estar en 2 modos:
- **Comando**. Permite modificar la celda (tipo de celda, crear o borrar celdas, etc). Resaltadas en *azul*.
- **Edición**. Permite teclear código o texto a formatear. Resaltadas en *verde*.

Al presionar ***shift+enter***:
1. Se ejecuta el código o realiza el formato de texto.
2. Termina la edición y pasa la celda al modo comando.
3. Salta a la siguiente celda.

```julia
1+2+3+4
```

## Celda de texto

texto, *texto*, texto! $A\subset B $

```julia
1+2+3
```

### Algunos comando útiles
- ***m*** y ***y***. Convierte la celda a tipo texto y a tipo código, resp.
- ***a*** y ***b***. Inserta celdas arriba y abajo, resp.
- ***dd*** y ***z***. Borra y desborra la celda, resp.
- ***enter*** y ***esc***. Entra en modo edición y en modo comando, resp.
